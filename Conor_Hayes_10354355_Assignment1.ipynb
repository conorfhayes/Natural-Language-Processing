Calculate the frequency (number of occurrences), *ignoring case*, of all words and their unigram probability from the corrected *training* sentences.

*Hint: use `Counter` to implement this so it may be called many times*

````
from collections import Counter

# gets the unigram of word, that is passed as parameter
def unigram(word):
    # calls to get lower case of the word
    a = multiplelowerCase(train_corrected)
    # implements count that counts all words in a
    word_counter = Counter(a)
    # uses counter to count the occurance of word
    unigram_word = word_counter[word]
    return unigram_word


def prob(word):
    # gets lower case of all in train_corrected and assigns to a
    a = multiplelowerCase(train_corrected)
    # freqdis uses nltk freqdist
    freqdist_samples = nltk.FreqDist(a).N()
    b = unigram(word)
    # reuturns the probability of the word 
    prob = float(b)/float(freqdist_samples)
    
    return prob

# Test your code with the following
e = unigram("me")
print("unigram('me')==", e)
````

## **Task 3** (15 Marks): 
[Edit distance](https://en.wikipedia.org/wiki/Edit_distance) is a method that calculates how similar two strings are to one another by counting the minimum number of operations required to transform one string into the other. There is a built-in implementation in NLTK that works as follows:


````
from nltk.metrics.distance import edit_distance
# Edit distance returns the number of changes to transform one word to another
print(edit_distance("hello", "hi"))
````

Write a function that calculates all words with *minimal* edit distance to the misspelled word. You should do this as follows

1. Collect the set of all unique tokens in `train`
2. Find the minimal edit distance, that is the lowest value for the function `edit_distance` between `token` and a word in `train`
3. Output all unique words in `train` that have this same (minimal) `edit_distance` value

*Do not implement edit distance, use the built-in NLTK function `edit_distance`*

````
# returns the candidates of a given word not in unique tokens
def get_candidates(token):

    holder = []
    corrected = {}
    corrected_sorted = {}

    for words in unique_words:
        token = token.lower()
        # get the edit distance of each word compared to the passed token
        edit_distance_num = edit_distance(token, words)
        # append each word and its minimum edit distance to a dictionary
        corrected.update({words: edit_distance_num})

    # sort the array holding all words and their edit distance from the token
    corrected_sorted_hold = sorted((value, key) for (key, value) in corrected.items())
    # once sorted the minimum edit distance will be equal to the value at location [0][0]
    minimum_edit_distance = (corrected_sorted_hold[0][0])
    for key, value in corrected_sorted_hold:
        # if the key is equal to the minimum edit distance append the value to the holder array
        
        if key == minimum_edit_distance:
          holder.append(value)
          
        # if not then we know we are no longer comparing the minimum edit distance therefore we return
        else:
          return holder
        
    return holder
````

## Task 4 (15 Marks):

Write a function that takes a (misspelled) sentence and returns the corrected version of that sentence. The system should scan the sentence for words that are not in the dictionary (set of unique words in the training set) and for each word that is not in the dictionary choose a word in the dictionary that has minimal edit distance and has the highest *unigram probability*. 

*Your solution to this should involve `get_candidates`*


````
def correct(sentence):
    sentence_in = [sentence]
    sentence_out = []
    word_candidates = []
    word_prob = {}
    correct_words = []
    prob_max = {}

    for words in sentence_in:
        for each_word in words:
            # if word is not in unique words list we assume it needs to be checked
            if each_word not in unique_words:
                # pass the word to get candidates function
                corrected_word = get_candidates(each_word)
                word_candidates.append(corrected_word)
            else:
                # else the word is correct and we append it to the corrected words array
                word_candidates.append(each_word)
                correct_words.append(each_word)

        for candidate in word_candidates:
          # if candiate is in the correct words array we add it to the sentence out
            if candidate in correct_words:
                sentence_out.append(candidate)
            else:
              # if not we look at all the candidates
                for word in candidate:
                  # for each candidate we get the prob of the word
                    prob_word = prob(word)
                    # we update each candidate and their corresponding prob to a dictionary
                    word_prob.update({word: prob_word})
                
                # we then sort this dictionary in reverse, putting max prob at location [0][0]
                prob_hold = sorted(word_prob.items(), key=operator.itemgetter(1), reverse=True)
                maximum_value = prob_hold[0][0]
                # max prob is then assign to value at location [0][0] and we get the word that 
                # corresponds to this value and append it to our sentence out array
                sentence_out.append(maximum_value)

            word_prob = {}
            prob_max = {}
    return sentence_out
````

````
# Import all libraries needed to complete question 6
import nltk
import operator
from collections import Counter
from nltk.metrics.distance import edit_distance
from nltk import pos_tag, ne_chunk, word_tokenize
import time
from nltk.corpus import wordnet
````

````
# correct_mod is the same as correct function, with a few minor additions
# which i will highlight with comments
def correct_mod(sentence):
    sentence_in = [sentence]
    sentence_out = []
    word_candidates = []
    word_prob = {}
    correct_words = []
    prob_max = {}

    for words in sentence_in:
        for each_word in words:
            each_word = each_word
            named_ents(each_word)

            if each_word in named_entitys:
                correct_words.append(each_word)
                word_candidates.append(each_word)
            
            elif each_word in unique_words:
                word_candidates.append(each_word)
                correct_words.append(each_word)

            elif each_word not in unique_words:
                corrected_word = get_candidates(each_word)
                word_candidates.append(corrected_word)

        for candidate in word_candidates:
            if candidate in correct_words:
                sentence_out.append(candidate)
            else:
                for word in candidate:
                    if len(sentence_out) != 0:
                        word_tense = tense(word)
                        # check to see if a past tense of the word exists
                        if word != word_tense:
                            # if it does, ensure this word is selected by assigning it 0.99 prob
                            word_prob.update({word: 0.99})
                        else:
                            # call the bigram prob for each word in candidates
                            word_1 = sentence_out[len(sentence_out)-1]
                            bigram_prob = bigram(word_1, word)
                            word_prob.update({word: bigram_prob})

                    else:
                        prob_word = prob(word)
                        word_prob.update({word: prob_word})
                
                # sort the array to isolate the max prob
                prob_hold = sorted(word_prob.items(), key=operator.itemgetter(1), reverse=True)
                maximum_value = prob_hold[0][0]
                sentence_out.append(maximum_value)

            word_prob = {}
            prob_max = {}
    return sentence_out
  
# Tense Detection Implementation Below
def tense(word):
    tag = pos_tag((word_tokenize(word)))

    for t in tag:
      # check if tag contains VBN, if so check if word + ed exist
      # if it does return this word
        if 'VBN' in t:
            word_ed = word + 'ed'
            if word_ed in unique_words:
                return word_ed
              
        # check if tag contains VBN, if so check if word + ed exist
        # if it does return this word
        if 'VBD' in t:
            word_ed = word + 'ed'
            if word_ed in unique_words:
                return word_ed

        else:
            return word
          
# Named Entity Detection Implementation Below also below is Add-One Smoothing Implementation
def named_ents(word):
    # word.capitalize is used to as named entitys are case sensitive
    tag = pos_tag((word_tokenize(word.capitalize())))
    tag_itv = pos_tag((word_tokenize(word)))
    # using ne_chunk to find named entitys
    ne_tree = ne_chunk(pos_tag(word_tokenize(word)))
    for tree in ne_tree:
        if hasattr(tree, 'label'):
          named_entitys.append(word)
    
    # detection for ITV
    for t in tag_itv:
        if 'I.T.V' in t:
            named_entitys.append(word)
    # if a parameter word is a valid word check the following
    if wordnet.synsets(word):
        for t in tag:
          # if cd append to named entity list
            if "CD" in t:
               named_entitys.append(word)
          # if NNS append to named entity list
            elif 'NNS' in t:
                named_entitys.append(word)
          # if NNP append to named entity list
            elif 'NNP' in t:
                named_entitys.append(word)

    return named_entitys
  
  
def bigram(word1, word2):
    # similar method like getting the unigram probabiity but we use twowords
    word1 = word1.lower()
    word2 = word2.lower()
    a = multiplelowerCase(train_corrected)
    # gather the bigrams of all the words in train_corrected
    bigrams = nltk.ngrams(a, 2)
    # count the bigrams
    bigrams_freq = Counter(bigrams)
    #get the freqdist of word1 and word2
    freq_holder = bigrams_freq[(word1, word2)]
    named_entitys_lower = []

    for each in named_entitys:
        named_entitys_lower.append(each.lower())

    word_counter = Counter(a)
    # if word one is a named entity get the unigram prob of word 2
    if word1 in named_entitys_lower:
        bigram_prob = prob(word2)
    else:
        word_count = word_counter[word1]
        bigram_prob = freq_holder / word_count

    # Add one smoothing, if the bigram prob = 0 perform add one smoothing
    if bigram_prob == 0:
        freq_holder = freq_holder + 1
        word_count = word_counter[word1] + len(unique_words)
        bigram_prob = freq_holder / word_count

    return bigram_prob
````
